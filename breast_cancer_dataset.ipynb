```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load dataset
data = load_breast_cancer()

# Convert to DataFrame for better visualization
df = pd.DataFrame(data.data, columns=data.feature_names)

# Add target column (0 = malignant, 1 = benign)
df['target'] = data.target

# Display first few rows
df.head()
```

```python
# Print number of features and samples
num_features = data.data.shape[1]
num_samples = data.data.shape[0]

print(f"Number of features: {num_features}")
print(f"Number of samples: {num_samples}")
```

```python
# Get unique class labels and their counts
num_classes = len(data.target_names)
class_counts = pd.Series(data.target).value_counts()

print(f"Number of classes: {num_classes}")
print(f"Class labels: {data.target_names.tolist()}")
print("Class distribution:")
print(class_counts.to_string())
```

```python
# Basic statistics of dataset
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot class distribution
sns.countplot(x=df['target'])
plt.xticks([0, 1], ['Malignant', 'Benign'])
plt.title('Class Distribution')
plt.show()
```
